### Find infected value for dst in China and not in China for each day

```python
in_table = folder + r'routesAverage_2.5.xlsx'
df = pd.read_excel(in_table)
print(len(df))
df = df.iloc[:,17:] 
# df.count(axis='columns')
df.head(3)
```

```python
# filter out the destination
cols = df.columns
for j in range(len(df.columns)):
    if j%2 == 1:
        col_name = cols[j]
        df[col_name] = df[col_name].str.slice(-3,).isin(list(cn_lst)) # export: stop=3; import: -3,
df.head(3)
```

```python
for i in range(8,101):
    print("Day " + str(i))
    col1 = str(i)+': Infected'
    col2 = str(i)+': Routes'
    cn_value = sum(df[df[col2] ==1][col1])
    ncn_value = sum(df[df[col2] ==0][col1].reset_index(drop=True).dropna())
    print(cn_value)
    print(ncn_value)
    print(cn_value+ncn_value)
```
