### city risk

```python
# filter out the destination
cols = df.columns
for j in range(len(df.columns)):
    if j%2 == 1:
        col_name = cols[j]
        df[col_name] = df[col_name].str.slice(-3,) # export: stop=3; import: -3,
df.head(3)
```

```python
# find column names
colist_infected = []
colist_dst = []
for j in range(len(df.columns)):
    col_name = cols[j]
    if j%2 == 0:
        colist_infected.append(col_name)
    if j%2 == 1:        
        colist_dst.append(col_name)
```

```python
# multiple cols to one cols
df_infected = df[colist_infected]
df_dst = df[colist_dst]
df_infected_one = pd.DataFrame({'infected':df_infected.values.ravel(order='F')})
df_dst_one = pd.DataFrame({'dst':df_dst.values.ravel(order='F')})
```

```python
# merge two cols
df_updated = pd.concat([df_infected_one, df_dst_one], axis=1, sort=False)
```

```python
## make sure the relationship remains
# result = df_updated
# output_csv = folder + 'output_TEST_' + str(cut) + '_'  + t.strftime('%m%d%y%H%M') + '.csv'
# result.to_csv(output_csv, index=False, encoding='utf-8')
```

```python
# groupby and sort
df_rank = df_updated.groupby('dst').sum().reset_index()
df_rank = df_rank.sort_values(by='infected',ascending = False).reset_index(drop=True)
df_rank.head(5)
```

```python
# import airport info
in_table = r'C:\Users\Ensheng\Desktop\Coronavirus\IATA\IATA_airports.xlsx'
df_iata = pd.read_excel(in_table)
print(len(df_iata))
df_iata.head(3)
```

```python
# make sure no one left behind
set(df_rank['dst']) - set(df_iata['Code'])
```

```python
# get airport info for the ranking list
result = pd.merge(df_rank, df_iata, how='left', left_on='dst',right_on='Code')
result
```

```python
# output
output_csv = folder + 'output_airport_' + str(cut) + '_'  + t.strftime('%m%d%y%H%M') + '.csv'
result.to_csv(output_csv, index=False, encoding='utf-8')
```

### country risk

```python
df_rank_country = result.groupby('Country Name')['infected'].sum().reset_index()
df_rank_country = df_rank_country.sort_values(by='infected',ascending = False).reset_index(drop=True)
df_rank_country.head(10)
```

```python
result = df_rank_country
output_csv = folder + 'output_country_' + str(cut) + '_'  + t.strftime('%m%d%y%H%M') + '.csv'
result.to_csv(output_csv, index=False, encoding='utf-8')
```
